###### What are the three key elements to keep in mind when you design an explanatory visualization?
1. **Start with a question**: before starting to design, you should know what you want to communicate to the audience. Look at the dataset, and decide what takeaways your visualization should have.

2. **Allow exploration**: visualizations are more engaging if they allow the audience to create their own experience of how they discover it. A good visualization should be interactive and give freedom to the one who is consuming it.   In order to achieve that you can follow this 3-step principle: "*Start by giving an overview*", "*Zoom and filter*" and finally "*Details on demand*".

3. **Know your readers**: it is very important to have in mind the target audience. You should be able to adapt your visualization depending on who you want to communicate to.

###### Explain in your own words: How is explanatory data analysis different from exploratory data analysis?

The main difference is that on one hand, **exploratory data** is used to **analyze data** and try to draw conclusions from the data behaviour we see. 

On the other hand, **explanatory data** is used at a later stage, once the observations have been made and we want to **present our findings** to an audience. 

They are complementary, since explanatory data is based on exploratory, and exploratory data needs explanatory to be communicated.
